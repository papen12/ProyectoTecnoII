# Notebook 09: Sistema de Alertas de Voz

Este notebook prueba de forma aislada el motor de síntesis de voz en español `pyttsx3` con cooldowns de repetición.

```python
import sys
import os
import time

sys.path.append(os.path.abspath('..'))
from services.voice_service import VoiceAlertService
```

```python
voice_service = VoiceAlertService()

print("Probando primera alerta (debe sonar)...")
voice_service.alert("Pare")

print("Probando segunda alerta del mismo elemento inmediatamente (debe ser bloqueada por el cooldown)...")
voice_service.alert("Pare")

print("Esperando 5 segundos...")
time.sleep(5)

print("Probando alerta después del cooldown (debe sonar)...")
voice_service.alert("Pare")

voice_service.shutdown()
```
