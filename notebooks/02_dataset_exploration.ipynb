# Notebook 02: Exploración de Estructura y Dataset

Este notebook se enfoca en verificar e inspeccionar la estructura del dataset propio exportado desde **Roboflow** en la carpeta `dataset/roboflow/`.

```python
import sys
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import cv2

sys.path.append(os.path.abspath('..'))
from configs import settings
```

```python
dataset_path = Path(settings.DATASET_DIR)
print(f"Ruta del dataset: {dataset_path.resolve()}")

if not dataset_path.exists():
    print("⚠️ ¡ATENCIÓN! La carpeta del dataset no existe. Asegúrate de colocar tu exportación de Roboflow en dataset/roboflow/")
else:
    # Listar subdirectorios para comprobar la estructura
    for child in dataset_path.iterdir():
        print(f"  - {child.name} ({'Directorio' if child.is_dir() else 'Archivo'})")
```

```python
# Mostrar la cantidad de imágenes en cada split si existen
for split in ['train', 'val', 'test']:
    split_img_dir = dataset_path / split / 'images'
    if split_img_dir.exists():
        count = len(list(split_img_dir.glob('*')))
        print(f"Cantidad de imágenes en {split}: {count}")
```
