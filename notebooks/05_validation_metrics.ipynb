# Notebook 05: Validación de Métricas

Este notebook se enfoca en evaluar el rendimiento de tu modelo entrenado `best.pt` analizando curvas mAP, precisión, recall y matriz de confusión.

```python
import sys
import os
from pathlib import Path

sys.path.append(os.path.abspath('..'))
from configs import settings
from services.model_loader import load_model
from utils.metrics import compute_metrics, generate_metrics_table
```

```python
model_path = Path(settings.MODELS_DIR) / "best.pt"
data_yaml_path = Path(settings.DATASET_DIR) / "data.yaml"

if not model_path.exists():
    print("⚠️ models/best.pt no encontrado. Debes realizar el entrenamiento primero (Notebook 04).")
elif not data_yaml_path.exists():
    print("⚠️ dataset/roboflow/data.yaml no encontrado.")
else:
    # Cargar modelo entrenado y validar
    model = load_model(model_path)
    metrics = compute_metrics(model, data_yaml_path)
    
    df_metrics = generate_metrics_table(metrics)
    print("\n--- TABLA DE MÉTRICAS OBTENIDAS ---")
    print(df_metrics)
```
