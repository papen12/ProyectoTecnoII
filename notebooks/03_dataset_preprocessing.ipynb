# Notebook 03: Validación del Dataset

Este notebook importa las funciones del módulo `utils/preprocessing.py` para asegurar que el dataset exportado desde Roboflow no contenga inconsistencias.

```python
import sys
import os
from pathlib import Path

sys.path.append(os.path.abspath('..'))
from configs import settings
from utils.preprocessing import validate_dataset, get_class_distribution
```

```python
dataset_dir = settings.DATASET_DIR
print(f"Validando dataset en: {dataset_dir}")

report = validate_dataset(dataset_dir)
for split, info in report.items():
    print(f"\nSplit: {split.upper()}")
    print(f"  - Total imágenes: {info['total_images']}")
    print(f"  - Total labels: {info['total_labels']}")
    if info['missing_labels']:
        print(f"  - ⚠️ Imágenes sin etiqueta ({len(info['missing_labels'])}): {info['missing_labels'][:5]}...")
    else:
        print("  - ✅ Integridad perfecta: Cada imagen tiene su archivo de labels.")
```

```python
# Cargar y graficar la distribución de clases en el dataset
df_dist = get_class_distribution(dataset_dir)
if not df_dist.empty:
    print("\nDistribución de instancias por Clase y Split:")
    print(df_dist)
else:
    print("⚠️ No se encontraron labels para calcular la distribución de clases.")
```
