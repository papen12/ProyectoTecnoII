# Notebook 06: Detección en Imágenes

Este notebook carga tu modelo entrenado `best.pt` y realiza predicciones estáticas en imágenes de prueba.

```python
import sys
import os
from pathlib import Path
import cv2
import matplotlib.pyplot as plt

sys.path.append(os.path.abspath('..'))
from configs import settings
from services.detector import TrafficSignDetector
from utils.drawing import draw_detections
```

```python
model_path = Path(settings.MODELS_DIR) / "best.pt"

if not model_path.exists():
    print("⚠️ models/best.pt no encontrado. Utilizando un modelo YOLO preentrenado para fines de prueba...")
    # Fallback temporal si no se ha entrenado aún
    detector = TrafficSignDetector(settings.MODEL_NAME)
else:
    detector = TrafficSignDetector(model_path)
```

```python
# Cambia la ruta a alguna imagen de prueba en tu dataset local
test_image_dir = Path(settings.DATASET_DIR) / "test" / "images"
images = list(test_image_dir.glob("*"))

if not images:
    print(f"⚠️ No se encontraron imágenes en {test_image_dir}.")
else:
    img_path = images[0]
    print(f"Inferencia sobre: {img_path.name}")
    
    frame = cv2.imread(str(img_path))
    detections = detector.detect(frame)
    
    # Dibujar predicciones
    result_frame = draw_detections(frame, detections)
    
    # Visualizar con matplotlib
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(result_frame, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```
